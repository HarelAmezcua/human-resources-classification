```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.preprocessing import OneHotEncoder
```

```python
base_dir= os.path.dirname(os.getcwd())
data_frame = pd.read_csv(os.path.join(base_dir, 'datasets', 'Cleaned_Human_Resources.csv'))
# Transform Attrition from Yes no to 0 and 1
data_frame['Attrition'] = data_frame['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
data_frame['OverTime'] = data_frame['OverTime'].apply(lambda x: 1 if x == 'Yes' else 0)
data_frame.head()
```

```python
encoder = OneHotEncoder()
```

```python
df_object = data_frame.select_dtypes(include=['object'])
X = encoder.fit_transform(df_object).toarray()
X
```

```python
aux_df = data_frame.drop(columns=df_object.columns, inplace=True)
aux_df = data_frame.drop(columns=['Attrition'])
aux_df = pd.concat([pd.DataFrame(X), aux_df], axis=1)
aux_df.columns = aux_df.columns.astype(str)
aux_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data_frame_scaled = scaler.fit_transform(aux_df)
data_frame_scaled = pd.DataFrame(data_frame_scaled, columns=aux_df.columns)
```
